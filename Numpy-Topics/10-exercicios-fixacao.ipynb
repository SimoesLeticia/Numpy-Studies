## Para os exercícios 7 e 8 considere o array a seguir.


|              | Cliente 01 | Cliente 02 | Cliente 03 |
|:------------:|:----------:|:----------:|:----------:|
| **Ano 01**   |      1     |      3     |      2     |
| **Ano 02**   |      0     |      1     |      0     |
| **Ano 03**   |      2     |      1     |      4     |
| **Ano 04**   |      0     |      0     |      0     |
| **Ano 05**   |      1     |      1     |      0     |

```python
# Os elementos representão a quantidade de acidentes cometidos pelos clientes a cada ano.
array_acidentes = np.array([[1, 3, 2],
                            [0, 1, 0],
                            [2, 1, 4],
                            [0, 0, 0],
                            [1, 1, 0]])
array_acidentes
```

## 07. Identifique a média de acidentes anual e a média de acidentes por cliente.

```python
# Encontra a média de acidentes por ano
media_acidentes_anual = np.mean(array_acidentes, axis=1)
media_acidentes_anual
```

```python
# Encontra a média de acidentes por cliente
media_acidentes_cliente = np.mean(array_acidentes, axis=0)
media_acidentes_cliente
```

## 08. Quais clientes passaram pelo menos 2 anos sem cometer acidentes?

```python
clientes_sem_acidentes = np.where(np.sum(array_acidentes == 0, axis=0) >= 2)[0]
clientes_sem_acidentes
```

## 15. Utilizando um índice booleano, crie um array contendo os dados da maior espécie encontrada *(Dica: esse valor corresponde a 22)*.

```python
# Encontra o maior valor da coluna Tamanho
maior_valor_colTamaho= np.max(especies[:, 3])

# Cria mask
indice_maior_especie = especies[:, 3] == maior_valor_colTamaho

# Aplica o mask ao array
maior_especie = especies[indice_maior_especie]
maior_especie
```

## 16. Através do índice sofisticado (fancy indexing), crie um array incluindo apenas os dados da espécie com ID 297.

```python
# Encontra indice do ID 297
indice_especie_297 = np.where(especies[:, 0] == 297)

# Crie um array incluindo apenas os dados da espécie com ID 297 usando fancy indexing
especie_297_dados = especies[indice_especie_297]
especie_297_dados
```

## 17. Utilizando np.where(), crie um array contendo apenas os dados da espécie cuja quantidade de representantes encontrados é 105.
